```python
import pandas as pd
import bs4
import requests
import openpyxl
```

```python
data = requests.get('https://www.trueid.net/th-th/search/%E0%B8%A1%E0%B8%B0%E0%B9%80%E0%B8%9F%E0%B8%B7%E0%B8%AD%E0%B8%87?tab=read')
soup = bs4.BeautifulSoup(data.text)
topic_list = []
detail_list = []
link_list = []
for n in soup.find_all('div' ,{'class' : 'style__Card-sc-cz5bs7-0 eoFeke'}):
    topic_list.append(n.find('div' ,{'class' : 'style__DescriptionWrapper-sc-cz5bs7-1 iYdQtP'}).find('h3').text)
    detail_list.append(n.find('h4',{'class' : 'style__Detail-sc-cz5bs7-4 dqaEoX'}).text)
    link_list.append(n.find('a',{'class' : 'style__Link-sc-1a2wk93-0 gkhKPw'}).get('href')) 
    
table = pd.DataFrame([topic_list, detail_list, link_list]).transpose()
table.columns = ['topic', 'detail', 'link']
table.set_index('topic')
```
